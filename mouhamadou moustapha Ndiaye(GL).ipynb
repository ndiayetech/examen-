```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#chargement des donnÃ©es
df = pd.read_csv("raw_data.csv")
df
```

```python
#vÃ©erifi...
df.isna().sum()
```

```python
#filtrer les valeurs manquants
masque = df.isna()
df[masque]
```

```python
#supprimer les valeurs maquants
df = df.dropna()
df
```

```python
df.columns
```

```python
df = df[['iso_code', 'location', 'date', 'total_cases', 'total_deaths',
       'stringency_index', 'population', 'gdp_per_capita',
       'human_development_index',]]
```

```python
df
```

```python
#CAS QUOTIDIENS DE COVID 19 DE 3 PAYS LES PLUS TOUCHES

#REGROUPER LES DONNEES PAR PAYS ET CALCULER LE NOMBRE TOTAL DE CAS
total_cas_par_pays = df.groupby('location')['total_cases'].max()

#Trier les pays par nombre total de cas et sÃ©lÃ©ctionner le 3 les plus touchÃ©s
pays_plus_touches = total_cas_par_pays.nlargest(3).index

#Filtrer les donnees pour inclure uniquement les 3 pays les plus touchÃ©s
donnee_pays_plus_touches = df[df["location"].isin(pays_plus_touches)]

#convertir la colonne date en type datetime
donnee_pays_plus_touches['date'] = pd.to_datetime(donnee_pays_plus_touches['date'])

#Tracer les cas quotidiens de covid 19 pour les 3 pays les plus touchÃ©s
plt.figure(figsize=(12, 8))

#parcourir les pays pour tracer chaque ligne et ajouter les lÃ©gendes
for pays in pays_plus_touches:
    donnees_pays = donnee_pays_plus_touches[donnee_pays_plus_touches['location'] == pays]
    plt.plot(donnees_pays['date'], donnees_pays['total_cases'], marker='o', linestyle='-', label=pays)
plt.title('Cas quotidiens de Covid 19 des 3 pays les plus touchÃ©s')
plt.xlabel('date')
plt.ylabel('Cas quotidien de covid-19')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#TAUX DE MORTALITE DES 3 PAYS LES PLUS TOUCHES

#Calculer le taux de mortalitÃ© pour chaque pays
donnee_pays_plus_touches['TAUX_MORTALITE'] = donnee_pays_plus_touches['total_deaths'] / donnee_pays_plus_touches['total_cases']*100

#Groupe par pays et calculer le taux de mortalite moyen
taux_mortalite_pays = donnee_pays_plus_touches.groupby('location')['TAUX_MORTALITE'].mean()

#Tracer le taux de mortalite en diagramme en barre
taux_mortalite_pays.plot(kind='bar', color='red')
plt.title('Taux de mortatÃ© des 3 pays les plus touchÃ©s par Covid 19')
plt.xlabel('Pays')
plt.ylabel('Taux de mortalitÃ©')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#EVOLUTION DU PIB DES 3 PAYS LES PLUS TOUCHES

#Filter les donnÃ©es pour les 3 pays le plus touches
top_3_pays = df.groupby('location')['total_cases'].max().nlargest(3).index
df_top_3 = df[df['location'].isin(top_3_pays)]

#Convertir le colonne date en type datetime
df_top_3['date'] = pd.to_datetime(df_top_3['date'])

#AgrÃ©rer les donnÃ©es par mois
df_top_3['MOIS'] = df_top_3['date'].dt.to_period('M')

#Calculer les indicateurs Ã©conomiques
df_top_3['PIB_total'] = df_top_3['gdp_per_capita'] * df_top_3['population']

#Tracer l'Ã©volution des iondicateurs Ã©conomiques et des statistiques Covid-19
plt.figure(figsize=(12, 8))

#Tracer l'Ã©volution du PIB TOTAL par mois pour chaque pays
for pays in top_3_pays:
    donnee_pays = df_top_3[df_top_3['location'] == pays]
    pib_total_mois = donnee_pays.groupby('MOIS')['PIB_total'].sum()
    plt.plot(pib_total_mois.index.astype(str), pib_total_mois, label=pays)

plt.title('Taux de mortatÃ© des 3 pays les plus touchÃ©s par Covid 19')
plt.xlabel('Mois')
plt.ylabel('PIB_total')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#CAS QUOTIDIENS DE COVID 19 DE 3 PAYS LES MOINS TOUCHES

#REGROUPER LES DONNEES PAR PAYS ET CALCULER LE NOMBRE TOTAL DE CAS
total_cas_par_pays = df.groupby('location')['total_cases'].min()

#Trier les pays par nombre total de cas et sÃ©lÃ©ctionner le 3 les plus touchÃ©s
pays_moins_touches = total_cas_par_pays.nlargest(3).index

#Filtrer les donnees pour inclure uniquement les 3 pays les plus touchÃ©s
donnee_pays_moins_touches = df[df["location"].isin(pays_moins_touches)]

#convertir la colonne date en type datetime
donnee_pays_moins_touches['date'] = pd.to_datetime(donnee_pays_moins_touches['date'])

#Tracer les cas quotidiens de covid 19 pour les 3 pays les plus touchÃ©s
plt.figure(figsize=(12, 8))

#parcourir les pays pour tracer chaque ligne et ajouter les lÃ©gendes
for pays in pays_moins_touches:
    donnees_pays = donnee_pays_moins_touches[donnee_pays_moins_touches['location'] == pays]
    plt.plot(donnees_pays['date'], donnees_pays['total_cases'], marker='o', linestyle='-', label=pays)
plt.title('Cas quotidiens de Covid 19 des 3 pays les moins touchÃ©s')
plt.xlabel('date')
plt.ylabel('Cas quotidien de covid-19')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

#TAUX DE MORTALITE DES 3 PAYS LES MOINS TOUCHES

#Calculer le taux de mortalitÃ© pour chaque pays
donnee_pays_moins_touches['TAUX_MORTALITE'] = donnee_pays_moins_touches['total_deaths'] / donnee_pays_moins_touches['total_cases']*100

#Groupe par pays et calculer le taux de mortalite moyen
taux_mortalite_pays = donnee_pays_moins_touches.groupby('location')['TAUX_MORTALITE'].mean()

#Tracer le taux de mortalite en diagramme en barre
taux_mortalite_pays.plot(kind='bar', color='yellow')
plt.title('Taux de mortalitÃ© des 3 pays les moins touchÃ©s par Covid 19')
plt.xlabel('Pays')
plt.ylabel('Taux de mortalitÃ©')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#EVOLUTION DU PIB DES 3 PAYS LES PLUS TOUCHES

#Filter les donnÃ©es pour les 3 pays le plus touches
min_3_pays = df.groupby('location')['total_cases'].min().nsmallest(3).index
df_top_3 = df[df['location'].isin(min_3_pays)]

#Convertir le colonne date en type datetime
df_top_3['date'] = pd.to_datetime(df_top_3['date'])

#AgrÃ©rer les donnÃ©es par mois
df_top_3['MOIS'] = df_top_3['date'].dt.to_period('M')

#Calculer les indicateurs Ã©conomiques
df_top_3['PIB_total'] = df_top_3['gdp_per_capita'] * df_top_3['population']

#Tracer l'Ã©volution des iondicateurs Ã©conomiques et des statistiques Covid-19
plt.figure(figsize=(12, 8))

#Tracer l'Ã©volution du PIB TOTAL par mois pour chaque pays
for pays in min_3_pays:
    donnee_pays = df_top_3[df_top_3['location'] == pays]
    pib_total_mois = donnee_pays.groupby('MOIS')['PIB_total'].sum()
    plt.plot(pib_total_mois.index.astype(str), pib_total_mois, label=pays)

plt.title('Taux de mortalitÃ© des 3 pays les moins touchÃ©s par Covid 19')
plt.xlabel('Mois')
plt.ylabel('PIB_total')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#L'INDICE DE SEVERITE DES MESURES PRISES DANS LES PAYS DE L'AFRIQUE DE L'OUEST

# Liste des pays de l'Afrique de l'Ouest
west_african_countries = ['Senegal', 'Gambia', 'Ghana', 'Nigeria', 'Mali', 'Burkina Faso', 'Cote d\'Ivoire', 'Liberia', 'Sierra Leone', 'Guinea', 'Guinea-Bissau', 'Cape Verde', 'Benin', 'Togo']

# Filtrer les donnÃ©es pour les pays de l'Afrique de l'Ouest
df_west_african_countries = df[df['location'].isin(west_african_countries)]

# Convertir la colonne de date en format de date
df_west_african_countries['date'] = pd.to_datetime(df_west_african_countries['date'])

# AgrÃ©ger les donnÃ©es par mois
df_west_african_countries['MOIS'] = df_west_african_countries['date'].dt.to_period('M')

# Calculer l'indice de sÃ©vÃ©ritÃ© moyen par mois
monthly_severity_index = df_west_african_countries.groupby('MOIS')['stringency_index'].mean()

# Tracer l'Ã©volution de l'indice de sÃ©vÃ©ritÃ© au fil du temps sous forme de diagramme en barres
plt.figure(figsize=(12, 8))
plt.bar(monthly_severity_index.index.astype(str), monthly_severity_index.values, color='skyblue')
plt.title('Indice de sÃ©vÃ©ritÃ© des mesures prises dans les pays de l\'Afrique de l\'Ouest')
plt.xlabel('Mois')
plt.ylabel('Indice de sÃ©vÃ©ritÃ©')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
